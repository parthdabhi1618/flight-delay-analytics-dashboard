```python
import pandas as pd

df = pd.read_csv("Airline_Delay_Cause.csv")

print(df.shape)
df.head()
```

```python
columns_needed = [
    "year",
    "month",
    "carrier",
    "carrier_name",
    "airport",
    "airport_name",
    "arr_flights",
    "arr_del15",
    "carrier_delay",
    "weather_delay",
    "nas_delay",
    "security_delay",
    "late_aircraft_delay"
]

df = df[columns_needed]
```

```python
df = df.fillna(0)
```

```python
df["total_delay_minutes"] = (
    df["carrier_delay"]
    + df["weather_delay"]
    + df["nas_delay"]
    + df["security_delay"]
    + df["late_aircraft_delay"]
)
```

```python
df["delay_rate"] = df["arr_del15"] / df["arr_flights"]
```

```python
df["delay_rate"] = df["delay_rate"].replace([float("inf"), -float("inf")], 0)
```

```python
import calendar

df["month_name"] = df["month"].apply(lambda x: calendar.month_name[int(x)])
```

```python
df.to_csv("flight_delay_cleaned.csv", index=False)
```
